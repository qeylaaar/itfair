```python
import pandas as pd
import re

# Path ke file data
PANEN_DATA_PATH = '../data/sample_data_panen.csv'
CUACA_DATA_PATH = '../data/sample_data_cuaca.csv'

# Membaca data panen dengan delimiter semicolon
df_panen = pd.read_csv(PANEN_DATA_PATH, sep=';')
print("Data Panen Mentah:")
print(df_panen.head())
print("\nTipe Data Panen Mentah:")
print(df_panen.info())

# Membaca data cuaca dengan delimiter semicolon
df_cuaca = pd.read_csv(CUACA_DATA_PATH, sep=';')
# Konversi kolom Tanggal ke datetime
df_cuaca['Tanggal'] = pd.to_datetime(df_cuaca['Tanggal'])
print("\nData Cuaca Mentah:")
print(df_cuaca.head())
print("\nTipe Data Cuaca Mentah:")
print(df_cuaca.info())
```
